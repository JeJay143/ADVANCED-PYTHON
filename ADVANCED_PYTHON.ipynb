**NUMPY**

```python
import numpy as np
print(np.__version__)
```

ARRAY CREATION

```python
# Array from a list
arr1 = np.array([1, 2, 6, 6])
print("Array from a list\n", arr1.astype(int))

# Array of zeros
arr2 = np.zeros((4, 4))
print("Array of zeros\n", arr2.astype(int))

# Array from ones
arr3 = np.ones((2, 2))
print("Array of ones\n", arr3.astype(int))

# Array with a range of values
arr4 = np.arange(0, 5)
print("Array within a range is\n", arr4.astype(int))

# Array from random values
arr5 = np.random.rand(3, 3)
print("Array from random values\n", (arr5 * 10).astype(int))  # Scaling and converting to integers
```

ARRAY ATTRIBUTES

```python
# Shape
arr1 = np.array([[1, 2, 6, 6], [3,5,7,9]])
print("The shape of arr1 is\n",arr1.shape)

print("It has data type\n",arr1.dtype)
```

SLICING AND INDEXING

```python
# Getting the forst element
arr1 = np.array([1, 2, 6, 6])
print("The first element is", arr1[0])

# Getting the last element
arr1 = np.array([3, 4, 5, 8])
print("The last element is", arr1[-1])

# Getting the last element
arr1 = np.array([3, 4, 5, 8])
print("The last element is", arr1[-1])

# Getting the second to last element
arr1 = np.array([3, 4, 5, 8])
print("The second to last element are", arr1[1:4])
```

ARRAY MANIPULATION

```python
# Reshape
list = np.array([[1, 2, 6, 6], [3,5,7,9]])
print('The new shape is\n',list.reshape(4, 2))

# Vertical Stack
list_stack = np.vstack([list, list])
print('The stacked version \n', list_stack)

# Statistical Operation
# Mean
list = np.array([[1, 2, 6, 6], [3,5,7,9]])
print("The mean of the list is\n",np.mean(list))

# Median
print("The median of the list is\n",np.median(list))

# Standard Deviation
print("The std of the list is\n",np.std(list))
```

```python
import pandas as pd
print(pd.__version__)
```

DATAFRAME CREATION

```python
# From a List
list = [2,2,1,4,5]
df = pd.DataFrame(list, columns=['column_name'])

# From a Dictionary
list = {"A": [3,1,5,3,2], 'B': [1,2,3,4,7]}
df1 = pd.DataFrame(list)
```

```python
df1[['A','B']]
```

DATA MANIPULATION

```python
# Adding a column
df1['C'] = pd.Series([7,8,9])
df1[['A','B', 'C']]
```

```python
# Deleting a row
df1.drop('C', axis = 1, inplace=True)
df1
```

```python
# Renaming a column
df1.rename(columns={'A': 'A_COL'}, inplace=True)
df1
```

DATA CLEANING

```python
# Checking for null data
my_list = [1,2,4, np.nan]
df = pd.DataFrame(my_list, columns=['column_name'])
df.isnull()
```

```python
  # Dropping null values
df.dropna(inplace=True)
```

```python
# Filling null values
df.fillna(value = 1, inplace = True)
df
```

```python
# Replacing Values
df.replace(1, 10, inplace=True)
df
```

GROUPING AGGREGATE

```python
# Group By
num = {'A': [1,2,3,4], 'C': [3,8,9,0]}
mun = pd.DataFrame(num)
mun.groupby('A')
```

```python
#Aggregation
mun.agg({'A': ['min', 'max', 'sum', 'mean']})
```

MERGING CONCATENATING AND JOINING

```python
# Concatenating
df1 = pd.DataFrame({'A': [1,3,2,4], 'B': [7,8,9,5]})
df2 = pd.DataFrame({'A': [7,8,9,0], 'B': [1,5,7,8]})
df = pd.concat([df1, df2])
df
```

```python
# Merging
df1 = pd.DataFrame({'A': [1,3,2,4], 'B': [7,8,9,5]})
df2 = pd.DataFrame({'A': [1,3,2,4], 'C': [11,15,12,8]})
df = pd.merge(df1, df2, on='A')
df
```

```python
# Joining
df1 = pd.DataFrame({'A': [1,2,3], 'B': [4,5,6]})
df2 = pd.DataFrame({'C': [7,8,9]})
df = df1.join(df2)
df

```

WORKING WITH DATES

```python
# Sample DataFrame with a date column
data = {'date_column': ['2022-01-15', '2023-02-20', '2024-03-25']}
df = pd.DataFrame(data)

# Convert 'date_column' to datetime type
df['date_column'] = pd.to_datetime(df['date_column'])

# Extract date parts
df['year'] = df['date_column'].dt.year
df['month'] = df['date_column'].dt.month
df['day'] = df['date_column'].dt.day

print(df)
```

FILE I/O

```python
from sqlalchemy import create_engine

# Sample data
data = {'Name': ['Alice', 'Bob', 'Charlie'],
        'Age': [25, 30, 35],
        'City': ['New York', 'Los Angeles', 'Chicago']}

# Create a DataFrame
df = pd.DataFrame(data)

# Write data to CSV file
df.to_csv('data.csv', index=False)

# Write data to Excel file
df.to_excel('data.xlsx', index=False)

# Write data to JSON file
df.to_json('data.json', orient='records')

# Write data to SQL database (SQLite in this example)
engine = create_engine('sqlite:///data.db')
df.to_sql('data_table', engine, index=False, if_exists='replace')

# Read data from CSV file
df_csv = pd.read_csv('data.csv')

# Read data from Excel file
df_excel = pd.read_excel('data.xlsx')

# Read data from JSON file
df_json = pd.read_json('data.json')

# Read data from SQL database
df_sql = pd.read_sql('data_table', engine)

# Display the data
print("Data from CSV file:")
print(df_csv)
print("\nData from Excel file:")
print(df_excel)
print("\nData from JSON file:")
print(df_json)
print("\nData from SQL database:")
print(df_sql)
```
